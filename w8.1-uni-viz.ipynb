```python
# %conda install seaborn 
# %conda install matplotlib 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/nkmwicz/data-for-students/refs/heads/main/soccer-players.csv"
df = pd.read_csv(url)
df.head()
```

```python
# Anatomy of a Viz
plt.figure(figsize=(5,3))
plt.xlabel("Height in CM")
plt.ylabel("Frequency")
plt.title("Histogram of Heights (cm) of Soccer Players")
plt.xlim(0,210)
plt.ylim(0,500)
plt.grid(True)
text = "skew: 0.12"
# plt.text(x=110, y=250, s=text, ha="center", va="center", fontsize=12, color="blue")
plt.text(
    x=0.05, 
    y=0.95, 
    s=text, 
    # ha="center", 
    va="center", 
    fontsize=12, 
    # color="blue",
    transform=plt.gca().transAxes)
plt.axvline(x=105, color="r", linestyle="--", label="Mean")

plt.axhline(y=250, color="b", linestyle="--", label="Center")
plt.legend()

plt.show()
```

```python
import scipy.stats as stats
```

```python
# Historigram in Plt
plt.figure(figsize=(5,3))
hts = df["height_in_cm"]
bins = round(hts.count() ** (1/3))
plt.hist(x=hts, bins=bins, edgecolor="black")
plt.xlabel("Height in CM")
plt.ylabel("Frequency")
plt.title("Histogram of Heights (cm) of Soccer Players")
plt.grid(True)
sk = stats.skew(hts, nan_policy="omit")
kurt = stats.kurtosis(hts, nan_policy="omit")
text = f"Skew: {sk:.3f}\nKurtosis: {kurt:.3f}"
# plt.text(x=110, y=250, s=text, ha="center", va="center", fontsize=12, color="blue")
plt.text(
    x=0.05, 
    y=0.90, 
    s=text, 
    # ha="center", 
    va="center", 
    fontsize=12, 
    # color="blue",
    transform=plt.gca().transAxes)
plt.axvline(x=hts.mean(), color="r", linestyle="--", label="Mean")
plt.axvline(x=hts.median(), color="g", linestyle="--", label="Median")
plt.legend()
```

```python
# Histogram in SNS
df1 = df.loc[df["height_in_cm"]>150]
plt.figure(figsize=(5,3))
sns.set_theme(style="whitegrid")
hist = sns.histplot(data=df1, x="height_in_cm", kde=True, line_kws={"color": "red"})
plt.xlabel("Height (cm)")
hts = df1["height_in_cm"]
sk = stats.skew(hts, nan_policy="omit")
kurt = stats.kurtosis(hts, nan_policy="omit")
text = f"Skew: {sk:.3f}\nKurtosis: {kurt:.3f}"
# plt.text(x=110, y=250, s=text, ha="center", va="center", fontsize=12, color="blue")
plt.text(
    x=0.05, 
    y=0.90, 
    s=text, 
    # ha="center", 
    va="center", 
    fontsize=12, 
    # color="blue",
    transform=plt.gca().transAxes)
plt.axvline(x=hts.mean(), color="r", linestyle="--", label="Mean")
plt.axvline(x=hts.median(), color="g", linestyle="--", label="Median")
plt.legend()
plt.show()
```
